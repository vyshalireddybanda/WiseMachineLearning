```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd  
from apyori import apriori  
```

```python
store_data = pd.read_csv('/content/drive/MyDrive/Wise ML/groceries.csv', header = None)  
store_data
```

```python
records = []  
for i in range(0, 9836):  
    records.append([str(store_data.values[i,j]) for j in range(0, 33)])
records
```

```python
association_rules = apriori(records, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)  

association_results = list(association_rules)

print(association_results)
```

```python
sorted(association_results, key=lambda x: x[2][0][3], reverse=True)
```

```python
for item in association_results:
    print([x for x in item[0]])
```

```python
for item in sorted(association_results, key=lambda x: x[2][0][3], reverse=True)[:5]:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0]  # frozenset:{light cream, chicken}
    items = [x for x in pair] #['light cream', 'chicken']
    print("Rule: " + items[0] + " -> " + items[1])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
