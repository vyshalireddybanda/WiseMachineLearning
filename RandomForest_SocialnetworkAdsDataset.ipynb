```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv("/content/drive/MyDrive/Wise ML/Social_Network_Ads.csv")
data.head()
```

```python
def convert(i):
  i = i.lower()
  if i=="male":
    return 0
  elif i=="female":
    return 1
  else:
    return -1
```

```python
data = pd.read_csv("/content/drive/MyDrive/Wise ML/Social_Network_Ads.csv",converters={'Gender':convert})
```

```python
data.shape
```

```python
data.head()
```

```python
data.columns
```

```python
labels = data["Purchased"]
features = data[['User ID',	'Gender',	'Age',	'EstimatedSalary']]
```

**Splitting data into train and test sets:**

```python
 X_train, X_test, y_train, y_test = train_test_split(features,labels, test_size=0.33, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

**Applying random forest:**

```python
clf = RandomForestClassifier(max_depth=2)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
data['Purchased'].value_counts()
```

```python
from sklearn.metrics import f1_score, classification_report
f1_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
